```python
import fileinput

print("Dobro došli u moj kviz!!!")

igra = input("Da li želiš igrati igru? ")

if igra.lower() != "da":
    quit()
else:
    print("U tom slučaju počnimo! :D ")

bodovi = 0


odgovor = input("Koja je godina osnutka Izraela? ")
if odgovor.lower() != "1948":
    print("Nije točno!")
else:
    print("Odgovor je točan!")
    bodovi += 1

odgovor = input("Koji je glavni grad Rusije? ")
if odgovor.lower() != "moskva":
    print("Nije točno!")
else:
    print("Odgovor je točan!")
    bodovi += 1

odgovor = input("Tko je nacrtaio Mona Lisu? ")
if odgovor.lower() != "leonardo da vinci":
    print("Nije točno!")
else:
    print("Odgovor je točan!")
    bodovi += 1

odgovor = input("Koja je trenutni američki predsjenik? ")
if odgovor.lower() == "joe biden":
    print("Odgovor je točan!")
    bodovi += 1
else:
    print("Nije točno")

if bodovi == 1:
    print (f"Skupio si ukupno {bodovi} bod.")
elif bodovi == 2 or 3 or 4:
    print(f"Skupio si ukupno {bodovi} boda.")
else:
    print(f"Skupio si ukupno {bodovi} bodova.")
```
