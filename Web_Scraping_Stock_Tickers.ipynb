```python
#This program scrapes stock tickers and their company name from a website
```

```python
#Import the dependencies
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#Create two empty lists for company name and the company ticker/symbol
company_name = []
company_symbol = []
```

```python
URL = "https://www.advfn.com/nyse/newyorkstockexchange.asp?companies=A"
page = requests.get(URL)
soup = BeautifulSoup(page.text, "html.parser")
odd_rows = soup.find_all("tr", attrs={"class":"ts0"})
even_rows = soup.find_all("tr", attrs={"class":"ts1"})
```

```python
for i in odd_rows:
  row = i.find_all("td")
  company_name.append(row[0].text.strip())
  company_symbol.append(row[1].text.strip())

for i in even_rows:
  row = i.find_all("td")
  company_name.append(row[0].text.strip())
  company_symbol.append(row[1].text.strip())
```

```python
#Create a function to scrape the data
def scrape_stock_symbols(Letter):
  Letter = Letter.upper()
  URL = "https://www.advfn.com/nyse/newyorkstockexchange.asp?companies="+Letter
  page = requests.get(URL)
  soup = BeautifulSoup(page.text, "html.parser")
  odd_rows = soup.find_all("tr", attrs={"class":"ts0"})
  even_rows = soup.find_all("tr", attrs={"class":"ts1"})

  for i in odd_rows:
    row = i.find_all("td")
    company_name.append(row[0].text.strip())
    company_symbol.append(row[1].text.strip())

  for i in even_rows:
    row = i.find_all("td")
    company_name.append(row[0].text.strip())
    company_symbol.append(row[1].text.strip())

  return(company_name, company_symbol)
```

```python
(temp_name, temp_symbol) = scrape_stock_symbols("b")
temp_symbol
```

```python
#Get the list of every letter in the alphabet
import string
string.ascii_uppercase
```

```python
#Loop through every letter in the alphabet to get all of the tickers and company names from the website
for char in string.ascii_uppercase:
  (temp_name, temp_symbol) = scrape_stock_symbols(char)
```

```python
#Create a dataFrame that contains the company name and company ticker
data = pd.DataFrame(columns=["company_name", "company_ticker"])
data["company_name"] = temp_name
data["company_ticker"] = temp_symbol

#Show data
data
```

```python
#Clean the data
data = data[data["company_name"] != ""]

#Show the data
data
```
